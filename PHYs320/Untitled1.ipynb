```python
def oneDimensionalWalk(N, ss):
    x = 0
    y = 0
    steps = N
    stepSize = ss;
    path=[0]
    for i in range(steps):
        if(random.random() < 0.5):
            x+=stepSize
        else:
            x-=stepSize
        path.append(x)
    return path
```

```python
finalSpots = []
walks = 1000
for i in range(walks):
    path = oneDimensionalWalk(100, 1)
    finalSpots.append(path[len(path)-1])
plt.hist(finalSpots, 60, facecolor='blue', alpha=0.5)
```

```python
def twoDimensionalWalk(N, ss):
    x = 0
    y = 0
    steps = N
    stepSize = ss;
    step = 1
    xPath = [0]
    yPath = [0]
    for i in range(steps):
        if(random.random() < 0.5):
            step = stepSize
        else:
            step = -stepSize
        if(random.random() < 0.5):
            x += step
        else:
            y += step
        xPath.append(x)
        yPath.append(y)
    return xPath, yPath
```

```python
xPath, yPath = twoDimensionalWalk(100, 1)
for i in range(len(xPath)-1):
    #plot([x1, x2], [y1, y2], color='k', linestyle='-', linewidth=2)
    plt.plot([xPath[i], xPath[i+1]], [yPath[i], yPath[i+1]], color='k', linestyle='-', linewidth=2)

    plt.scatter(0, 0, color="red", s=200)
    plt.scatter(xPath[-1], yPath[-1], color="blue", s = 200)
print("Red is where the walk started")
print("Blue is where the walk ended")
```
