```python
G=6.67*10**-11
Me=5.972*10**24
Mm=7.348*10**22
Re=6371
```

Since this is a gravity problem, above I have set 3 constants to be used later. The constants include the Gravitational Constant (G), the mass of the earth (Me), and the mass of the moon (Mm).

```python
vMoonX=0
vMoonY=-2
xMoon=363104+(Re/2.0)
yMoon=0
xEarth=0
yEarth=0
moonListX=[0]*365
moonListY=[0]*365
```

Next we setup some initial values that are not constant. These include the initial position of the moon, relative to the earth, and the initial velocity of the moon. In this case we are ignoring the gravitational force the moon exerts on the moon, and considering the center of the earth as the center of the orbit.

```python
def calcAcc(vectVal, rMag):
    f=lambda rMag: (G*Me*vectVal)/((rMag**3)+1e-15)
    return f(rMag)
```

```python
dayOfSeconds = 60*60*24
for i in range(0, 365):
    rMag = ((xMoon-xEarth)**2+(yMoon-yEarth)**2)*(1/2)
    Ax=calcAcc(xMoon-xEarth, rMag)
    Ay=calcAcc(yMoon-yEarth, rMag)
    
    vMoonX += Ax*dayOfSeconds
    vMoonY += Ay*dayOfSeconds
    xMoon+= vMoonX*dayOfSeconds
    yMoon += vMoonY*dayOfSeconds
    moonListX[i] = xMoon
    moonListY[i] = yMoon
    plt.xlabel("X Position")
    plt.ylabel("Radius")
    plt.plot(xMoon, rMag, "oc")
    #print(xMoon, yMoon)
```

```python
import matplotlib.pyplot as plt

plt.plot(moonListX, moonListY)
```
